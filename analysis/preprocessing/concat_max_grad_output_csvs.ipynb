```python
import pandas as pd
import numpy as np
import os
np.random.seed(0)
from typing import List
```

```python
DATA_DIR = "../../data/lgd_sva/"
# MODEL_ID = "bf-google/bert_uncased_L-6_H-128_A-2"
MODEL_ID = "bf-google/bert_uncased_L-6_H-512_A-8"

ENTROPY_EVAL_RESULTS_PATH = os.path.join(DATA_DIR, MODEL_ID, "results", "max_grad")
ENTROPY_EVAL_RESULTS_PATH = os.path.join(DATA_DIR, MODEL_ID, "results", "max_grad", "oplus")
AGG_FILENAME = "agg_max_grad_results.csv"
OUT_DATA_PATH = os.path.join(ENTROPY_EVAL_RESULTS_PATH, "agg_max_grad_results.csv")
```

```python
csv_paths = [x for x in os.listdir(ENTROPY_EVAL_RESULTS_PATH) if x[-3:] == "csv" and x != AGG_FILENAME and "max_grad" in x]
```

```python
dfs: List[pd.DataFrame] = []
for csv_path in csv_paths:
    df = pd.read_csv(
        os.path.join(ENTROPY_EVAL_RESULTS_PATH, csv_path), 
        names=[
            "row_num",
            "sentence_index",
            "is_correct",
            "num_distractors",
            "correct_word",
            "correct_word_logit",
            "incorrect_word",
            "incorrect_word_logit",
            "token",
            "token_index",
            "layer_num",
            "path_name",
            "path grad",
            "masked_sentence",
        ]
    )
    # df.drop(columns=["row_num", "smth", "smthelse"])
    dfs.append(df)
```

```python
agg_df = pd.concat(dfs, axis=0)
```

```python
agg_df = agg_df.drop(columns=["row_num"])
```

```python
agg_df = agg_df.reset_index().drop(columns=["index"])
```

```python
agg_df.to_csv(OUT_DATA_PATH, index=False)
```
